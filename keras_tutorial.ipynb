```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler

class DataTrainer:
  def __init__(self):
    self.trainLabels = []
    self.trainSamples = []
    self.scaledTrainSamples = None
    self.isConvertedToNumpy = False
  def appendRandIntSampleData(self, minInt, maxInt, labelValue):
    randVal = randint(minInt, maxInt)
    self.trainSamples.append(randVal)
    self.trainLabels.append(labelValue)
  def convertDataToNumpyArrays(self):
    if not self.isConvertedToNumpy:
      self.trainLabels = np.array(self.trainLabels)
      self.trainSamples = np.array(self.trainSamples)
      self.trainLabels, self.trainSamples = shuffle(self.trainLabels, self.trainSamples)
      self.isConvertedToNumpy = True
  def createScaledTrainSamples(self):
    if self.isConvertedToNumpy:
      scaler = MinMaxScaler(feature_range = (0, 1))
      self.scaledTrainSamples = scaler.fit_transform(self.trainSamples.reshape(-1, 1))
  def printSamples(self):
    for i in self.trainSamples:
      print(i)
  def printLabels(self):
    for i in self.trainLabels:
      print(i)
  def printScaledTrainSamples(self):
    if self.scaledTrainSamples is not None:
      for i in self.scaledTrainSamples:
        print(i)

dataTrainer = DataTrainer()

for i in range(50):
  dataTrainer.appendRandIntSampleData(13, 64, 1)
  dataTrainer.appendRandIntSampleData(65, 100, 0)

for i in range(1000):
  dataTrainer.appendRandIntSampleData(13, 64, 0)
  dataTrainer.appendRandIntSampleData(65, 100, 1)

dataTrainer.convertDataToNumpyArrays()
dataTrainer.createScaledTrainSamples()

dataTrainer.printScaledTrainSamples()
```
